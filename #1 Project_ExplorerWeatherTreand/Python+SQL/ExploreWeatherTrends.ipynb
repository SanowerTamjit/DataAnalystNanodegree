*Task 2: Show Line chart for city vs global temp*

```python
#importing libraries
import matplotlib.pyplot as plt
import pandas as panda
```

```python
#Reading the data
dhakaTemp = panda.read_csv('Desktop/UdacityDataAnalyst/city_data.csv')
globalTemp = panda.read_csv('Desktop/UdacityDataAnalyst/global_data.csv')
```

```python
#Moving Average Calculation (considering 6 years)
movAvgDhaka = dhakaTemp['avg_temp'].rolling(6).mean()
movAvgGlobal = globalTemp['avg_temp'].rolling(6).mean()
```

```python
print(movAvgDhaka)
print(movAvgGlobal)
```

```python
#Dhaka temperature Line Chart
plt.plot(dhakaTemp['year'],movAvgDhaka,label='Dhaka')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temp (C)") 
plt.title("Dhaka Temperature")
plt.show()
#Global Temparature Line Chart
plt.plot(globalTemp['year'],movAvgGlobal,label='Global')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temp (C)") 
plt.title("Global Temperature")
plt.show()
```

```python
#Comparison Chart between Dhaka and Global
plt.plot(dhakaTemp['year'],movAvgDhaka,label='Dhaka')
plt.plot(globalTemp['year'],movAvgGlobal,label='Global')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temp (C)") 
plt.title("Dhaka vs Global Temperature")
plt.show()
```
